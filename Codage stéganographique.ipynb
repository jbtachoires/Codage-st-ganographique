```python
from imageio import imread
import matplotlib.pyplot as plt

tf = imread(chem_png_texte)
print(tf)
```

```python
b=tf%4
a =b.dot([64,16,4,1])# on convertit l'image en  liste 
print(a)
```

```python

def escargot(a):
    r = []
    for i in range(len(a)):# on itère sur la première ligne de longueur différente
        r.append(a[i][0])


    for indice in range(len(a[0])//2):#on itère sur le nombre de spirales à effectuer
    
        for i in range(indice+1,len(a)-indice):
            r.append(a[len(a)-1-indice][i])
            
        ld = list(range(indice+1, len(a)-indice-1))
        ld.reverse()
        for i in ld:
            r.append(a[i][len(a[0])-1-indice])

        lg = list(range(indice+1,len(a)-indice))
        lg.reverse()
        for i in lg:
            r.append(a[indice][i])

        for i in range(indice+1,len(a)-2*(indice+1)+1+indice):
            r.append(a[i][indice+1])
    return(r)
r = escargot(a)
        
#fonction de test pour visualiser la spirale
tab_test = np.array( [[[i,j] for j in range(16)] for i in range(16)] )
L_test = escargot(tab_test)
plt.plot( [p[1] for p in L_test], [p[0] for p in L_test], ".-", label="parcours")
plt.grid() ; plt.axis("equal") ;
  
    
```
